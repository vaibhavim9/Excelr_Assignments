```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Q1 :

```python
company_name = ['Allied Signal',
                'Bankers Trust',
                'General Mills',
                'ITT Industries',
                'J.P.Morgan & Co.',
                'Lehman Brothers',
                'Marriott','MCI',
                'Merrill Lynch','Microsoft',
                'Morgan Stanley','Sun Microsystems',
                'Travelers',
                'US Airways',
                'Warner-Lambert']
measure_x=[24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.7,35]
```

```python
# plotting data using Pie Plot
colors = sns.color_palette('bright')[0:10]
plt.figure(figsize=(6,8))
plt.pie(measure_x,labels=company_name,colors = colors,radius=1,autopct='%1.0f%%')
plt.show()
```

```python
# plotting data using barplot
plt.bar(measure_x,company_name,width=2)
plt.show()
```

```python
#plotting data using boxplot
sns.boxplot(measure_x)
```

```python
#Histogram
sns.histplot(measure_x)
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x=company_name,y=measure_x)
```
